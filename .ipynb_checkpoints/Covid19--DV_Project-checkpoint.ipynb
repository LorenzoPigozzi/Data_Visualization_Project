### 3. Data cleaning <a class="anchor" id="3"></a>

```python
# dropping out the NaN for new_deaths
df_clean = df[~df['new_deaths'].isna()]
```

```python
# converting types of the columns

# date type
df['date'] = pd.to_datetime(df['date'], format='%Y/%m/%d')

# cases and deaths from float to integer
feats = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths']
df_clean[feats] = df_clean[feats].astype(int)
```

```python
# dropping negative values for the variables new_cases and new_deaths
df = df[~df['new_cases'] < 0]
df = df[~df['new_deaths'] < 0]
```

```python
# copying the result to the original data
df = df_clean.copy()
```

### 4. Lineplot <a class="anchor" id="4"></a>
#### Data transformations and plots
In order to conduct the analysis, it's necessary a data manipulation. The dataset should be transformed.

```python
europe_countries = ['Albania', 'Andorra', 'Austria', 'Belarus', 'Belgium',
       'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia',
       'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece',
       'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kosovo', 'Latvia',
       'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova',
       'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia', 'Norway',
       'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Serbia',
       'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine',
       'United Kingdom']
```

```python
## Manipulation + Plot
def lineplot_covid(variable_of_interest, list_of_countries):
    ##data manipulation##
    # seleting the variables useful for the plot 
    variables_for_plot = ['location', 'date', variable_of_interest]
    # creating a new dataset for the plot with the selected varaibles
    df_for_plot = df[variables_for_plot]

    # manipulating the data with pivot table
    df_for_plot = pd.pivot_table(df_for_plot, values=[variable_of_interest],columns=['location'], index = ['date'])
    # dropping the multi-index level for the columns
    df_for_plot.columns = df_for_plot.columns.droplevel(0)
    
    ##plotting##
    # creating the object data for the plot
    data_1 = [dict(type='scatter',
                 x=df_for_plot.index,
                 y=df_for_plot[location],
                 name=location)
                                for location in list_of_countries]
    # setting the layout
    plot_1_layout = dict(title=dict(text='Evolution of Covid in Europe: '+ variable_of_interest),
                      xaxis=dict(title='Date', 
                                rangeslider=dict(visible=True)),
                      yaxis=dict(title='Deaths per million')
                      )
    # displaying the graph
    plot_1 = go.Figure(data=data_1, layout=plot_1_layout)
    # plot_1.show(renderer='png')
    plot_1.show()
```

```python
lineplot_covid('new_cases', europe_countries)
```

```python
lineplot_covid('new_cases_per_million', ['Germany', 'Italy', 'Netherlands'])
```

```python
lineplot_covid('total_deaths_per_million', ['Germany', 'Italy', 'France'])
```

```python
lineplot_covid('new_deaths_per_million', ['Germany', 'Italy', 'Netherlands', 'France'])
```
